```python
import os
from langchain_core.messages import HumanMessage, SystemMessage, AIMessageChunk
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent

# ========== CONFIGURAГѓвЂЎГѓЖ’O ==========
os.environ["OPENAI_API_KEY"] = "OPENAI_API_KEY_AQUI"

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
llm_tradutor = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)


# ========== FERRAMENTAS ==========
@tool
def calculadora(expressao: str) -> str:
    """Ferramenta para realizar cГѓВЎlculos matemГѓВЎticos."""
    try:
        resultado = eval(expressao, {"__builtins__": {}}, {})
        return f"Resultado: {resultado}"
    except Exception as e:
        return f"Erro no cГѓВЎlculo: {str(e)}"

@tool
def tradutor(texto: str, idioma_destino: str = "inglГѓВЄs") -> str:
    """Ferramenta para traduzir textos."""
    try:
        prompt = f"Traduza o seguinte texto para {idioma_destino}. Retorne APENAS a traduГѓВ§ГѓВЈo:\n\n{texto}"
        response = llm_tradutor.invoke(prompt)
        return f"TraduГѓВ§ГѓВЈo: {response.content.strip()}"
    except Exception as e:
        return f"Erro na traduГѓВ§ГѓВЈo: {str(e)}"


# ========== PROMPT DO AGENTE ==========
AGENT_PROMPT = """
VocГѓВЄ ГѓВ© um assistente especializado que atua de forma explicativa e didГѓВЎtica, com ferramentas de cГѓВЎlculo e traduГѓВ§ГѓВЈo.

REGRA OBRIGATГѓвЂњRIA DE FORMATAГѓвЂЎГѓЖ’O:
Toda resposta DEVE seguir este formato exato:

[ExplicaГѓВ§ГѓВЈo detalhada do raciocГѓВ­nio, contexto e processo]

[Linha em branco]

[Resultado final claro e direto]

Exemplos:

Pergunta: "Quanto ГѓВ© 1 + 1 * 2?"
Resposta correta:
Esse cГѓВЎlculo contГѓВ©m uma soma e uma multiplicaГѓВ§ГѓВЈo. Na matemГѓВЎtica, seguimos a ordem de precedГѓВЄncia das operaГѓВ§ГѓВµes, onde a multiplicaГѓВ§ГѓВЈo vem antes da adiГѓВ§ГѓВЈo. Primeiro calculamos 1 * 2 = 2, e depois somamos 1 ao resultado.

O resultado de 1 + 1 * 2 ГѓВ© igual a 3.

---

Pergunta: "Traduza 'Bom dia' para francГѓВЄs"
Resposta correta:
A expressГѓВЈo "Bom dia" ГѓВ© uma saudaГѓВ§ГѓВЈo matinal comum em portuguГѓВЄs. Em francГѓВЄs, a saudaГѓВ§ГѓВЈo equivalente combina "bon" (bom) com "jour" (dia).

A traduГѓВ§ГѓВЈo de "Bom dia" para francГѓВЄs ГѓВ© "Bonjour".

---

SEMPRE siga esse padrГѓВЈo: explicaГѓВ§ГѓВЈo primeiro, linha em branco, depois resultado final.
"""


# ========== CRIAГѓвЂЎГѓЖ’O DO AGENTE ==========
agent = create_react_agent(llm, [calculadora, tradutor])


# ========== FUNГѓвЂЎГѓвЂўES ==========
def run_agent(user_input: str) -> str:
    """Executa o agente e retorna a resposta final."""
    messages = [
        SystemMessage(content=AGENT_PROMPT),
        HumanMessage(content=user_input)
    ]
    result = agent.invoke({"messages": messages})
    return result["messages"][-1].content


def run_agent_stream(user_input: str) -> str:
    """Executa o agente com streaming palavra por palavra."""
    messages = [
        SystemMessage(content=AGENT_PROMPT),
        HumanMessage(content=user_input)
    ]
    
    resposta_final = ""
    
    for event in agent.stream({"messages": messages}, stream_mode="messages"):
        msg, metadata = event
        
        if isinstance(msg, AIMessageChunk) and msg.content:
            print(msg.content, end="", flush=True)
            resposta_final += msg.content
    
    print()
    return resposta_final
```
