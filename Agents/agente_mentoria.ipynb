```python
# 3. DEFINIÃ‡ÃƒO DAS FERRAMENTAS (TOOLS)
# ----------------------------------------------------------------------------

@tool
def calculadora(expressao: str) -> str:
    """
    Ferramenta para realizar cÃ¡lculos matemÃ¡ticos.
    
    Args:
        expressao: ExpressÃ£o matemÃ¡tica como string (ex: "(2+3)*4/2")
    
    Returns:
        Resultado do cÃ¡lculo como string
    """
    try:
        # Avalia a expressÃ£o matemÃ¡tica de forma segura
        resultado = eval(expressao, {"__builtins__": {}}, {})
        return f"Resultado: {resultado}"
    except Exception as e:
        return f"Erro no cÃ¡lculo: {str(e)}"


@tool
def tradutor(texto: str, idioma_destino: str = "inglÃªs") -> str:
    """
    Ferramenta para traduzir textos entre portuguÃªs e inglÃªs.
    
    Args:
        texto: Texto a ser traduzido
        idioma_destino: Idioma de destino ("inglÃªs" ou "portuguÃªs")
    
    Returns:
        Texto traduzido
    """
    # DicionÃ¡rio simples de traduÃ§Ãµes (pode ser expandido ou usar API real)
    traducoes_pt_en = {
        "bom dia": "good morning",
        "boa tarde": "good afternoon",
        "boa noite": "good night",
        "olÃ¡": "hello",
        "tchau": "bye",
        "obrigado": "thank you",
        "por favor": "please",
    }
    
    traducoes_en_pt = {v: k for k, v in traducoes_pt_en.items()}
    
    texto_lower = texto.lower().strip()
    
    if idioma_destino.lower() == "inglÃªs":
        return traducoes_pt_en.get(texto_lower, f"[TraduÃ§Ã£o via LLM]: {texto} â†’ (usar API de traduÃ§Ã£o real)")
    elif idioma_destino.lower() == "portuguÃªs":
        return traducoes_en_pt.get(texto_lower, f"[TraduÃ§Ã£o via LLM]: {texto} â†’ (usar API de traduÃ§Ã£o real)")
    else:
        return "Idioma nÃ£o suportado. Use 'inglÃªs' ou 'portuguÃªs'."


# Lista de ferramentas disponÃ­veis
tools = [calculadora, tradutor]

# ----------------------------------------------------------------------------
```

```python
# 4. CONFIGURAÃ‡ÃƒO DO MODELO E BINDING DAS FERRAMENTAS
# ----------------------------------------------------------------------------

# Inicializa o modelo LLM
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# Vincula as ferramentas ao modelo
llm_with_tools = llm.bind_tools(tools)

# ----------------------------------------------------------------------------
# 5. DEFINIÃ‡ÃƒO DO ESTADO DO GRAFO
# ----------------------------------------------------------------------------

class AgentState(TypedDict):
    """Estado que serÃ¡ compartilhado entre os nÃ³s do grafo"""
    messages: Annotated[Sequence[BaseMessage], add_messages]
    contexto: str
    explicacao: str
    resposta_final: str


# ----------------------------------------------------------------------------
```

```python
# 6. FUNÃ‡Ã•ES DOS NÃ“S DO GRAFO
# ----------------------------------------------------------------------------

def agente_node(state: AgentState):
    """
    NÃ³ principal do agente - decide se precisa usar ferramentas ou responder.
    """
    messages = state["messages"]
    response = llm_with_tools.invoke(messages)
    return {"messages": [response]}


def formatar_resposta_node(state: AgentState):
    """
    NÃ³ responsÃ¡vel por formatar a resposta final no formato estruturado.
    """
    messages = state["messages"]
    
    # Cria prompt para formataÃ§Ã£o estruturada
    prompt_formatacao = f"""
Com base na conversa anterior, formate a resposta em trÃªs parÃ¡grafos seguidos, sem tÃ­tulos, marcadores ou tags:

Primeiro parÃ¡grafo (contexto): Descreva o que foi entendido da pergunta e quais ferramentas foram usadas, se houver.

Segundo parÃ¡grafo (explicaÃ§Ã£o): Explique o raciocÃ­nio e o passo a passo de como chegou Ã  resposta.

Terceiro parÃ¡grafo (resposta): Entregue a resposta final de forma clara e direta.

IMPORTANTE: 
- NÃ£o use tÃ­tulos como "Contexto:", "ExplicaÃ§Ã£o:", "Resposta:"
- NÃ£o use marcadores BEGIN/END
- Apenas escreva trÃªs parÃ¡grafos naturais separados por linha em branco
- Seja conciso mas completo
- NÃ£o use fÃ³rmulas LaTeX (\\( \\)), escreva matematicamente de forma simples
"""
    
    messages_com_prompt = list(messages) + [HumanMessage(content=prompt_formatacao)]
    resposta_formatada = llm.invoke(messages_com_prompt)
    
    return {"messages": [resposta_formatada], "resposta_final": resposta_formatada.content}


def should_continue(state: AgentState):
    """
    FunÃ§Ã£o de roteamento - decide se deve continuar usando ferramentas ou finalizar.
    """
    messages = state["messages"]
    last_message = messages[-1]
    
    # Se a Ãºltima mensagem tem tool_calls, continua para as ferramentas
    if hasattr(last_message, "tool_calls") and last_message.tool_calls:
        return "continue"
    # Caso contrÃ¡rio, vai para formataÃ§Ã£o
    return "format"


# ----------------------------------------------------------------------------
```

```python
# 7. CONSTRUÃ‡ÃƒO DO GRAFO
# ----------------------------------------------------------------------------

# IMPORTANTE: Recriar o grafo com a nova funÃ§Ã£o
workflow = StateGraph(AgentState)

workflow.add_node("agente", agente_node)
workflow.add_node("ferramentas", ToolNode(tools))
workflow.add_node("formatar", formatar_resposta_node)  # Usa a nova funÃ§Ã£o

workflow.set_entry_point("agente")

workflow.add_conditional_edges(
    "agente",
    should_continue,
    {
        "continue": "ferramentas",
        "format": "formatar"
    }
)

workflow.add_edge("ferramentas", "agente")
workflow.add_edge("formatar", END)

# RECOMPILA o grafo
app = workflow.compile()

# ----------------------------------------------------------------------------
# 8. FUNÃ‡ÃƒO PRINCIPAL DE EXECUÃ‡ÃƒO
```

```python
# 8. FUNÃ‡ÃƒO PRINCIPAL DE EXECUÃ‡ÃƒO
# ----------------------------------------------------------------------------

def run_agent(pergunta: str, verbose: bool = True):
    """
    Executa o agente com a pergunta fornecida.
    
    Args:
        pergunta: Pergunta do usuÃ¡rio
        verbose: Se True, exibe o processo de execuÃ§Ã£o
    
    Returns:
        Resposta formatada do agente
    """
    # Estado inicial
    initial_state = {
        "messages": [HumanMessage(content=pergunta)],
        "contexto": "",
        "explicacao": "",
        "resposta_final": ""
    }
    
    if verbose:
        print("ðŸ¤– Processando sua pergunta...\n")
    
    # Executa o grafo
    result = app.invoke(initial_state)
    
    # Extrai a resposta final
    resposta = result["resposta_final"]
    
    if verbose:
        print(resposta)
        print()
    
    return resposta


# ----------------------------------------------------------------------------
```

```python
# 9. VALIDAÃ‡ÃƒO DE FORMATO
# ----------------------------------------------------------------------------

def validar_formato(resposta: str) -> bool:
    """
    Valida se a resposta contÃ©m trÃªs parÃ¡grafos separados.
    """
    # Remove espaÃ§os extras e divide por linhas vazias
    paragrafos = [p.strip() for p in resposta.split('\n\n') if p.strip()]
    # Verifica se tem pelo menos 2 parÃ¡grafos (formato mÃ­nimo aceitÃ¡vel)
    return len(paragrafos) >= 2


# ----------------------------------------------------------------------------
```
