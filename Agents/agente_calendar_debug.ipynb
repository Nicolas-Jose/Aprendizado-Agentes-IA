```python
import os
from langchain_core.messages import HumanMessage, SystemMessage, AIMessageChunk
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain.agents import create_agent

os.environ["OPENAI_API_KEY"] = "OPENAI_API_KEY_AQUI"

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
llm_tradutor = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)

@tool
def calculadora(expressao: str) -> str:
    """Ferramenta para realizar cÃ¡lculos matemÃ¡ticos."""
    try:
        resultado = eval(expressao, {"__builtins__": {}}, {})
        return f"Resultado: {resultado}"
    except Exception as e:
        return f"Erro no cÃ¡lculo: {str(e)}"

@tool
def tradutor_tool(texto: str, idioma_destino: str = "inglÃªs") -> str:
    """Ferramenta para traduzir textos usando LLM interno."""
    try:
        prompt = f"Traduza o seguinte texto para {idioma_destino}. Retorne APENAS a traduÃ§Ã£o:\n\n{texto}"
        response = llm_tradutor.invoke(prompt)
        return f"TraduÃ§Ã£o: {response.content.strip()}"
    except Exception as e:
        return f"Erro na traduÃ§Ã£o: {str(e)}"

CALCULADORA_PROMPT = """
VocÃª Ã© um assistente especializado em matemÃ¡tica e cÃ¡lculos.

REGRA CRÃTICA: VocÃª DEVE SEMPRE usar a ferramenta 'calculadora' para realizar qualquer operaÃ§Ã£o matemÃ¡tica.
NUNCA calcule mentalmente. SEMPRE chame a ferramenta, mesmo para operaÃ§Ãµes simples como 2+2.

Processo obrigatÃ³rio:
1. Identifique a expressÃ£o matemÃ¡tica
2. Use a ferramenta calculadora passando a expressÃ£o exata (ex: "15 + 27")
3. Depois que receber o resultado da ferramenta, forneÃ§a uma explicaÃ§Ã£o didÃ¡tica
4. Apresente o resultado final

FORMATO DA RESPOSTA:
[ExplicaÃ§Ã£o didÃ¡tica do raciocÃ­nio matemÃ¡tico]
[Linha em branco]
[Resultado final obtido da ferramenta]

Exemplo de uso correto:
UsuÃ¡rio: "Quanto Ã© 5 + 3?"
VocÃª: [chama calculadora com "5 + 3"]
VocÃª: "Esta Ã© uma adiÃ§Ã£o simples. Somamos 5 unidades com 3 unidades.

O resultado de 5 + 3 Ã© igual a 8."
"""

calculadora_agent = create_agent(
    llm,
    tools=[calculadora],
    system_prompt=CALCULADORA_PROMPT,
    tool_choice="required"  # ForÃ§a o uso de tool
)

TRADUTOR_PROMPT = """
VocÃª Ã© um assistente especializado em traduÃ§Ã£o de idiomas.
REGRA OBRIGATÃ“RIA DE FORMATAÃ‡ÃƒO:
Toda resposta DEVE seguir este formato exato:
[ExplicaÃ§Ã£o sobre a expressÃ£o/palavra, contexto cultural ou linguÃ­stico]
[Linha em branco]
[Resultado da traduÃ§Ã£o claro e direto]
SEMPRE use a ferramenta de traduÃ§Ã£o e SEMPRE forneÃ§a contexto antes do resultado.
"""

tradutor_agent = create_agent(
    llm_tradutor,
    tools=[tradutor_tool],
    system_prompt=TRADUTOR_PROMPT
)

@tool
def realizar_calculo(solicitacao: str) -> str:
    """Realiza cÃ¡lculos matemÃ¡ticos com explicaÃ§Ã£o didÃ¡tica."""
    print("\n" + "  " + "ðŸ”§ SUB-AGENT: Calculadora Agent iniciado")
    print("  " + "â”€" * 76)
    
    # Usa stream para ver as etapas internas
    sub_step = 0
    for step in calculadora_agent.stream({
        "messages": [{"role": "user", "content": solicitacao}]
    }):
        sub_step += 1
        for node_name, update in step.items():
            print(f"\n  ðŸ”¹ Sub-Step {sub_step} - Node: {node_name}")
            
            for message in update.get("messages", []):
                # Indenta a saÃ­da do sub-agente
                lines = str(message.pretty_print).split('\n')
                message.pretty_print()
    
    result = calculadora_agent.invoke({
        "messages": [{"role": "user", "content": solicitacao}]
    })
    
    print("  " + "â”€" * 76)
    print("  " + "ðŸ”§ SUB-AGENT: Calculadora Agent concluÃ­do\n")
    
    return result["messages"][-1].content

@tool
def realizar_traducao(solicitacao: str) -> str:
    """Traduz textos entre idiomas com contexto explicativo."""
    print("\n" + "  " + "ðŸ”§ SUB-AGENT: Tradutor Agent iniciado")
    print("  " + "â”€" * 76)
    
    # Usa stream para ver as etapas internas
    sub_step = 0
    for step in tradutor_agent.stream({
        "messages": [{"role": "user", "content": solicitacao}]
    }):
        sub_step += 1
        for node_name, update in step.items():
            print(f"\n  ðŸ”¹ Sub-Step {sub_step} - Node: {node_name}")
            
            for message in update.get("messages", []):
                message.pretty_print()
    
    result = tradutor_agent.invoke({
        "messages": [{"role": "user", "content": solicitacao}]
    })
    
    print("  " + "â”€" * 76)
    print("  " + "ðŸ”§ SUB-AGENT: Tradutor Agent concluÃ­do\n")
    
    return result["messages"][-1].content

SUPERVISOR_PROMPT = """
VocÃª Ã© um assistente inteligente que coordena especialistas em matemÃ¡tica e traduÃ§Ã£o.
VocÃª tem acesso a dois especialistas:
1. Especialista em CÃ¡lculos - para todas as operaÃ§Ãµes matemÃ¡ticas
2. Especialista em TraduÃ§Ã£o - para traduzir textos entre idiomas
IMPORTANTE:
- Quando o usuÃ¡rio pedir cÃ¡lculos, use o especialista em cÃ¡lculos
- Quando o usuÃ¡rio pedir traduÃ§Ãµes, use o especialista em traduÃ§Ã£o
- Se a solicitaÃ§Ã£o envolver ambos, use os dois em sequÃªncia
- Sempre retorne a resposta completa dos especialistas
"""

supervisor_agent = create_agent(
    llm,
    tools=[realizar_calculo, realizar_traducao],
    system_prompt=SUPERVISOR_PROMPT
)

# ============================================================================
# FUNÃ‡ÃƒO DE TESTE COM LOGS DETALHADOS
# ============================================================================

def test_agent_with_detailed_logs(user_input: str):
    """
    Executa o agent mostrando TODAS as etapas:
    - Mensagens AI do supervisor
    - Tool calls do supervisor
    - Processos INTERNOS dos sub-agents (tool calls deles tambÃ©m!)
    - Resultados das tools
    - Resposta final
    """
    print("\n" + "="*80)
    print(f"ðŸ§ª TESTE: {user_input}")
    print("="*80 + "\n")
    
    step_count = 0
    
    for step in supervisor_agent.stream(
        {"messages": [{"role": "user", "content": user_input}]}
    ):
        step_count += 1
        print(f"\n{'â”€'*80}")
        print(f"ðŸ“Š SUPERVISOR STEP {step_count}")
        print(f"{'â”€'*80}")
        
        for node_name, update in step.items():
            print(f"\nðŸ“ Supervisor Node: {node_name}")
            
            # Processa as mensagens no update
            for message in update.get("messages", []):
                message.pretty_print()
                print()
    
    print("\n" + "="*80)
    print("âœ… TESTE CONCLUÃDO")
    print("="*80 + "\n")


# ============================================================================
# TESTES
# ============================================================================

if __name__ == "__main__":
    
    # TESTE 1: Apenas cÃ¡lculo simples
    print("\n" + "ðŸŽ¯ TESTE 1: CÃ¡lculo Simples")
    test_agent_with_detailed_logs("Quanto Ã© 15 + 27?")
    
    # TESTE 2: Apenas traduÃ§Ã£o
    print("\n" + "ðŸŽ¯ TESTE 2: TraduÃ§Ã£o Simples")
    test_agent_with_detailed_logs("Traduza 'hello world' para portuguÃªs")
    
    # TESTE 3: CÃ¡lculo mais complexo
    print("\n" + "ðŸŽ¯ TESTE 3: CÃ¡lculo Complexo")
    test_agent_with_detailed_logs("Calcule (25 * 4) + (100 / 5)")
    
    # TESTE 4: MÃºltiplas operaÃ§Ãµes (cÃ¡lculo + traduÃ§Ã£o)
    print("\n" + "ðŸŽ¯ TESTE 4: CÃ¡lculo + TraduÃ§Ã£o")
    test_agent_with_detailed_logs(
        "Calcule 50 * 3 e depois traduza o resultado para inglÃªs"
    )
    
    # TESTE 5: TraduÃ§Ã£o para idioma especÃ­fico
    print("\n" + "ðŸŽ¯ TESTE 5: TraduÃ§Ã£o para FrancÃªs")
    test_agent_with_detailed_logs("Traduza 'Bom dia' para francÃªs")
```
