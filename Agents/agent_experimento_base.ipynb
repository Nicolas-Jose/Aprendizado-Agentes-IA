```python
import os
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain.agents import create_agent
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
```

```python
os.environ["OPENAI_API_KEY"] = "OPENAI_API_KEY_AQUI"

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
llm_tradutor = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)
```

```python
AGENT_PROMPT = """VocÃª Ã© um assistente especializado em cÃ¡lculos e traduÃ§Ãµes.

ðŸŽ¯ REGRAS DE FORMATAÃ‡ÃƒO:
1. Seja claro e direto
2. Use emojis para categorizar aÃ§Ãµes:
   - ðŸ§® Para cÃ¡lculos matemÃ¡ticos
   - ðŸŒ Para traduÃ§Ãµes
   - ðŸ’¡ Para dicas ou sugestÃµes
3. Sempre explique qual ferramenta vai usar ANTES de usÃ¡-la
4. Formate nÃºmeros grandes com separadores (ex: 1.000.000)
5. Finalize perguntando se precisa de mais ajuda

EXEMPLO DE RESPOSTA IDEAL:
"ðŸ§® Vou calcular essa expressÃ£o para vocÃª!
[usa calculadora]
O resultado Ã© 150. 
ðŸ’¡ Esse valor representa... [contexto se necessÃ¡rio]
Posso ajudar com mais alguma coisa?"
"""
```

```python
@tool
def calculadora(expressao: str) -> str:
    """Ferramenta para realizar cÃ¡lculos matemÃ¡ticos."""
    try:
        resultado = eval(expressao, {"__builtins__": {}}, {})
        return f"Resultado: {resultado}"
    except Exception as e:
        return f"Erro no cÃ¡lculo: {str(e)}"


@tool
def tradutor(texto: str, idioma_destino: str = "inglÃªs") -> str:
    """Ferramenta para traduzir textos usando LLM interno."""
    try:
        prompt = f"Traduza o seguinte texto para {idioma_destino}. Retorne APENAS a traduÃ§Ã£o:\n\n{texto}"
        response = llm_tradutor.invoke(prompt)
        return f"TraduÃ§Ã£o: {response.content.strip()}"
    except Exception as e:
        return f"Erro na traduÃ§Ã£o: {str(e)}"


tools = [calculadora, tradutor]
```

```python
agent = create_agent(llm, tools)
```

```python
def run_agent_streaming(pergunta: str):
    """Executa o agente com streaming habilitado."""
    print("ðŸ¤– Processando sua pergunta...\n")
    
    resposta_completa = ""
    
    for chunk in agent.stream({"messages": [HumanMessage(content=pergunta)]}):
        for key, value in chunk.items():
            if key == "agent":
                for message in value.get("messages", []):
                    if hasattr(message, "content") and message.content:
                        texto = message.content
                        print(texto, end="", flush=True)
                        resposta_completa += texto
            
            elif key == "tools":
                for message in value.get("messages", []):
                    if hasattr(message, "content"):
                        print(f"\n[Tool: {message.name}] {message.content}")
    
    print("\n")
    return resposta_completa


def run_agent_simples(pergunta: str):
    """Executa o agente sem streaming (modo tradicional)."""
    print("ðŸ¤– Processando sua pergunta...\n")
    
    result = agent.invoke({"messages": [HumanMessage(content=pergunta)]})
    resposta = result["messages"][-1].content
    
    print(resposta)
    print()
    
    return resposta
```

```python
run_agent_streaming("Calcule 10+5 e traduza 'olÃ¡' para inglÃªs");
```

```python
run_agent_simples("Calcule 10+5 e traduza 'olÃ¡' para inglÃªs");
```
